````python
from faker import Faker
import uuid
import random
from datetime import datetime

fake = Faker('vi_VN')  # Tiếng Việt

def generate_sample_customer():
    gender = random.choice(['male', 'female'])
    name = fake.name_male() if gender == 'male' else fake.name_female()
    
    return {
        "id": str(uuid.uuid4()),
        "provider": random.choice(["messenger", "web"]),
        "account_id": str(fake.random_number(digits=15, fix_len=True)),
        "account_name": fake.user_name(),
        "avatar": fake.image_url(width=200, height=200),
        "fullname": name,
        "gender": gender,
        "birthday": fake.date_of_birth(minimum_age=20, maximum_age=40).strftime('%Y-%m-%d'),
        "phone": fake.phone_number(),
        "email": fake.email(),
        "address": fake.address().replace("\n", ", "),
        "last_message_at": fake.date_time_this_decade(before_now=True, after_now=False).strftime('%Y-%m-%d %H:%M:%S'),
        "last_message": fake.sentence(nb_words=8),
        "message_count": fake.random_int(min=1, max=5),
        "sentiment": random.choice(["negative", "neutral", "positive"]),
        "assigned_to": random.choice(["ai", "me"]),
    }

# Tạo danh sách 10 khách hàng
sample_customers = [generate_sample_customer() for _ in range(100)]

# In ra kết quả
import json
with open('customers.json', 'w', encoding='utf-8') as f:
    json.dump(sample_customers, f, ensure_ascii=False, indent=2)
````
